The central limit theorem (CLT) states that the distribution of a sample variable  approximates a normal distribution. 

Put anothe rway, CLT is a statistical premise that, given a sufficiently large sample size from a population with t finite level of variance, the mean of all sampled variables from the sample population will be approximately equal to the mean of the whole population. 

```python
import numpy as np  
import matplotlib.pyplot as plt
```

Here, we draw samples from a uniform distribution and calculate the sample mean and sample standard deviation. They follow the standard deviation. 

```python
means = []
stds = []
for i in range(5000): 
    samples = np.random.uniform(0, 5, size = 1000)
    mean = samples.mean()
    std = samples.std()
    means.append(mean)
    stds.append(std)

plt.title("Distribution of Sample Mean")
plt.hist(means)
plt.show()

plt.title("Distribution of Sample Standard Deviation")
plt.hist(stds)
plt.show()
```

Now we draw samples from an exponential distribution and calculate the sample mean and standard deviation. 

```python
means = []
stds = []
for i in range(5000): 
    samples = np.random.exponential(2, size = 1000)
    mean = samples.mean()
    std = samples.std()
    means.append(mean)
    stds.append(std)

plt.title("Distribution of Sample Mean")
plt.hist(means)
plt.show()

plt.title("Distribution of Sample Standard Deviation")
plt.hist(stds)
plt.show()
```
